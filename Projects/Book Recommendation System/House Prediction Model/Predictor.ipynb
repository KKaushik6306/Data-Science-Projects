### Data Cleansing

```python
data.isnull().sum()
```

```python
data['location'].value_counts()
```

```python
data['size']
```

```python
#Fill Null values of Location, size
data['location'] = data['location'].fillna('Sarjapur Road')
data['size'] = data['size'].fillna('2 BHK')
```

```python
data.isnull().sum()
```

```python
#Fill Null values of Bath & Balcony
data['bath'] = data['bath'].fillna(data['bath'].median())
```

```python
data['bhk'] = data['size'].apply(lambda x: x.split()[0])
```

```python
data
```

```python
data['bhk'].value_counts()
```

```python
data['total_sqft'].unique()
#We can see that in total sqft there are few range values, so we need to get the mean of those values and get it impute.
```

```python
def convert_range(x):
    temp = x.split('-')
    if len(temp)>1:
        return (float(temp[0]) + float(temp[0]))/2
    try:
        return float(x)
    except:
        return None
```

```python
data['total_sqft'] = data['total_sqft'].apply(convert_range)
```

```python
data.info()
```

```python
data.isnull().sum()
```

```python
#Creating New colum i.e. Price Per Square Feet
data['price_per_sqft'] = data['price'] * 100000 / data['total_sqft']
```

```python
data
```

```python
data.describe()
```

```python
#Removing extra spaces from location column
data['location'] = data['location'].apply(lambda x: x.strip())
```

```python
location_Names = data['location'].value_counts()
```

```python
location_count_lessthan10 = location_Names[location_Names<10]
```

```python
data['location'] = data['location'].apply(lambda x: 'other' if x in location_count_lessthan10 else x)
```
